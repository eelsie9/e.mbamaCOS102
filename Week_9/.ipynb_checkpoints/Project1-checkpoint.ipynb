```python
import tkinter as tk
from tkinter import messagebox

retail_bank = ["Lines of credit", "Investment management and accounts", "Insurance", 
        "Retirement and education accounts", "loans and mortgages", "Checking and saving"]

global_bank = ["Multi-currency management services and products", "Foreign currency accounts", "Foreign currency credit cards", 
          "Transborder advisory services", "Liquidity management"]

comm_bank = ["Lines of credit", "Investment management and accounts", "Insurance", "Advisory services"]

class zenith():

    def __init__(self, name):
        self.name = name

    def unique_services(self):
        return []

    def mutual_services(self):
        return ["Lines of credit", "Investment management and accounts", "Insurance"]

class retail_bank(zenith):
    
    def unique_services(self):
        return ["Retirement and education accounts", "loans and mortgages", "Checking and saving"]

class global_bank(zenith):
    
    def unique_services(self):
        return ["Multi-currency management services and products", "Foreign currency accounts", "Foreign currency credit cards", 
          "Transborder advisory services", "Liquidity management"]
    
    def mutual_services(self):
        return []

class comm_bank(zenith):

    def unique_services(self):
        return ["Advisory services"]

# GUI application class
class zenith_app:
    def __init__(self, root):
        self.root = root
        self.root.title("Zenith Bank Service Display")

        # Labels and entries
        tk.Label(root, text="Employee Name:").grid(row=0, column=0, padx=10, pady=5, sticky="e")
        self.name_entry = tk.Entry(root, width=30)
        self.name_entry.grid(row=0, column=1, pady=5)

        tk.Label(root, text="Division:").grid(row=1, column=0, padx=10, pady=5, sticky="e")
        self.division_var = tk.StringVar(value="Retail Banking")
        self.division_menu = tk.OptionMenu(root, self.division_var, "Retail Banking", "Global Banking", "Commercial Banking")
        self.division_menu.grid(row=1, column=1, pady=5)

        self.submit_button = tk.Button(root, text="Show Services", command=self.show_services)
        self.submit_button.grid(row=2, column=0, columnspan=2, pady=10)

        self.output_text = tk.Text(root, height=10, width=60)
        self.output_text.grid(row=3, column=0, columnspan=2, padx=10, pady=10)

    def show_services(self):
        name = self.name_entry.get().strip()
        division = self.division_var.get()

        if not name:
            messagebox.showerror("Input Error", "Please enter the employee name.")
            return

        # Determine the correct subclass to use
        if division == "Retail Banking":
            employee = retail_bank(name)
        elif division == "Global Banking":
            employee = global_bank(name)
        elif division == "Commercial Banking":
            employee = comm_bank(name)
        else:
            messagebox.showerror("Error", "Invalid division selected.")
            return

        # Compose output
        output = f"Employee Name: {employee.name}\nDivision: {division}\n\n"
        services = employee.mutual_services() + employee.unique_services()

        output += "Services Rendered:\n"
        for service in services:
            output += f" - {service}\n"

        # Display result
        self.output_text.delete("1.0", tk.END)
        self.output_text.insert(tk.END, output)

if __name__ == "__main__":
    root = tk.Tk()
    app = zenith_app(root)
    root.mainloop()
```
