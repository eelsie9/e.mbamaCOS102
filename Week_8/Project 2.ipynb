```python
import tkinter as tk
from tkinter import messagebox

class Delivery():

    def __init__(self, location, weight):
        self.location = location
        self.weight = weight

    def calculate(self):
        if self.location == "PAU":
            if self.weight >= 10:
                return 2000
            else:
                return 1500
        elif self.location == "Epe":
            if self.weight >= 10:
                return 5000
            else:
                return 4000
        else:
            return None

def check_price():
    location = location_var.get()
    try:
        weight = float(weight_entry.get())
    except ValueError:
        messagebox.showerror("Invalid Input", "Please enter a valid number for weight.")
        return

    delivery = Delivery(location, weight)
    charge = delivery.calculate()

    if charge is not None:
        messagebox.showinfo("Delivery Charge", f"Total charge: â‚¦{charge}")
    else:
        messagebox.showwarning("Error", "Invalid location selected.")

window = tk.Tk()
window.title("Delivery Charge Calculator")
window.geometry("350x250")

# Labels and input
tk.Label(window, text="Select Location:", font=("Arial", 12)).pack(pady=5)
location_var = tk.StringVar(window)
location_var.set("PAU")  # Default value
tk.OptionMenu(window, location_var, "PAU", "Epe").pack(pady=5)

tk.Label(window, text="Enter Package Weight (kg):", font=("Arial", 12)).pack(pady=5)
weight_entry = tk.Entry(window)
weight_entry.pack(pady=5)

# Button
tk.Button(window, text="Calculate Charge", command=check_price).pack(pady=20)

window.mainloop()
    
```
